```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['international_plan'] = le.fit_transform(df['international_plan'])
df['voice_mail_plan'] = le.fit_transform(df['voice_mail_plan'])
df['churn'] = le.fit_transform(df['churn'])
```

```python
df.head()
```

```python
df.columns
```

```python
df['minutes'] = df['total_day_minutes'] + df['total_eve_minutes'] + df['total_night_minutes'] +df['total_intl_minutes']
```

```python
df['calls'] = df['total_day_calls'] + df['total_eve_calls'] + df['total_night_calls'] + df['total_intl_calls']
```

```python
df['charges'] = df['total_day_charge'] + df['total_eve_charge'] + df['total_night_charge'] +df['total_intl_charge']
```

```python
x = df[['account_length','international_plan','voice_mail_plan','number_vmail_messages','minutes','calls','charges','number_customer_service_calls']]
```

```python
y = df['churn']
```

```python
x.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_scaled = scaler.fit_transform(x)
```

```python
x_scaled = pd.DataFrame(x_scaled)
```

```python
x_scaled
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression()
```

```python
classifier.fit(x_scaled,y)
```

```python
test = pd.read_csv('test.csv')
```

```python
test.head()
```

```python
test['minutes'] = test['total_day_minutes'] + test['total_eve_minutes'] + test['total_night_minutes'] +test['total_intl_minutes']
test['calls'] = test['total_day_calls'] + test['total_eve_calls'] + test['total_night_calls'] + test['total_intl_calls']
test['charges'] = test['total_day_charge'] + test['total_eve_charge'] + test['total_night_charge'] +test['total_intl_charge']
x_test = test[['account_length','international_plan','voice_mail_plan','number_vmail_messages','minutes','calls','charges','number_customer_service_calls']]
```

```python
x_test.head()
```

```python
le = LabelEncoder()
x_test['international_plan'] = le.fit_transform(x_test['international_plan'])
x_test['voice_mail_plan'] = le.fit_transform(x_test['voice_mail_plan'])
```

```python
x_test.head()
```

```python
x_scaled_test = scaler.fit_transform(x_test)
```

```python
x_scaled_test = pd.DataFrame(x_scaled_test)
```

```python
x_scaled_test.head()
```

```python
result = classifier.predict(x_scaled_test)
```

```python
result =pd.DataFrame(result)
```
